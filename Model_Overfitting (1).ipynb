# ***Fitting the model and evaluating***

```python
maxdepths =[2,3,4,5,6,7,8,9,10,15,20,25,30,35,40,45,50,55]
```

```python
trainAcc = np.zeros(len(maxdepths))
```

```python
trainAcc
```

```python
testAcc = np.zeros(len(maxdepths))
```

```python
testAcc
```

```python
index = 0
```

```python
for depth in maxdepths:
  clf=tree.DecisionTreeClassifier(max_depth=depth)
  clf=clf.fit(X_train,y_train)
  y_predicTrain=clf.predict(X_train)
  y_predicTest=clf.predict(X_test)
  trainAcc[index] = accuracy_score(y_train,y_predicTrain)
  testAcc[index] = accuracy_score(y_test,y_predicTest)
  index += 1
```

```python
testAcc
```

```python
import matplotlib.pyplot as plt
import matplotlib_inline
```

```python
plt.plot(maxdepths,trainAcc, 'ro-', maxdepths,testAcc, 'bv--')
plt.legend(['Training Accuracy', 'Testing Accuracy'])
plt.xlabel('Max Depth')
plt.ylabel('Accuracy')
```
